```python
root = 'https://www.rottentomatoes.com/browse/movies_in_theaters/'

title = []
release_date = []
sentimen = []

for page in range(1, 5):
  url = requests.get(root + str(page))
  soup = BeautifulSoup(url.text, 'html.parser')

  for q in soup.find_all('div', {'class':'flex-container'}):
    title.append(q.find("span", {"class":"p--small"}).text)

  for a in soup.find_all('div', {'class':'flex-container'}):
    release_date.append(a.find("span", {"class":"smaller"}).text)

  for t in soup.find_all('div', {'class':'flex-container'}):
    sentimen.append(t.find("rt-text", {"slot":"criticsScore"}).text)
```

```python
import pandas as pd
df = pd.DataFrame(
    {'Titles':title,
     'Release Date':release_date,
     'Critics Score':sentimen
     })
df.head(11)
```
